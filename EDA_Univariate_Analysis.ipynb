## Part 1: Understanding Product Categories
**Business Question**: What are the most popular product categories on Amazon UK, and how do they compare in terms of listing frequency?

```python
#frequency table 
category_count= df['category'].value_counts()
category_freq= df['category'].value_counts(normalize=True)
```

```python
catedory_freq_table=pd.concat([category_count,category_freq], axis=1).head(5).reset_index()
catedory_freq_table
```

```python
# **Visualizations**

sns.barplot(data=catedory_freq_table, x='category', y='count',palette='coolwarm')
plt.title('Top 5 of most popular product categories on Amazon UK')
plt.xticks(rotation=45, ha="right")

plt.show()
```

```python
category_freq.head(5).plot.pie(autopct='%1.1f%%', pctdistance=0.8, labeldistance=1.2, colors=sns.color_palette('Set2'))
plt.title('Top 5 of most popular product categories on Amazon UK')
plt.show()
```

### Part 2: Delving into Product Pricing
**Business Question**: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

```python
# 1. **Measures of Centrality**:
price_mean=round(df['price'].mean(),2)
price_median=round(df['price'].median(),2)
price_mode=df['price'].mode()

print(f"the medium price is {price_mean}")
print(f"the median price is {price_median}")
print(f"the mode price is {price_mode}")
```

```python
# 2. **Measures of Dispersion**:
# the variance, standard deviation, range, and interquartile range
price_mean=round(df['price'].mean(),2)
price_median=round(df['price'].median(),2)
price_mode=df['price'].mode()

#calculate interquartile range of values in the 'points' column
q75, q25 = np.percentile(df['price'], [75 ,25])
price_iqr = q75 - q25

print(f"the average rate is {price_mean}")
print(f"the median rate is {price_median}")
print(f"the mode rate is {price_mode}")
print(f"the IQR is {price_iqr}")
```

```python
# 3. **Visualizations**:
sns.displot(df, x="price", bins=100)
plt.show()
```

```python
sns.boxplot(df['price'], orient="h")
plt.show()
```

```python
df_price_filter=df[df['price']<60000]
```

```python
sns.boxplot(df_price_filter['price'], orient="h")
plt.xlim(0, 100)
plt.show()
```
